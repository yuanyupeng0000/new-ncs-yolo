```python
#get target images
import json
import os
import cv2
targetJason = "/media/yyp/DATA/Public_Datasets/Generated/\
selected_cls_COCO_train_2017.json"
nameStr = []
with open(targetJason, "r+") as f:
    data = json.load(f)
    print("read ready")
for i in data:
    imgName = str(i["filename"]) + ".jpg"
    nameStr.append(imgName)
nameStr = set(nameStr)
nameStr = sorted(nameStr)
print(nameStr)
print(len(nameStr))
```

```python
path = '/media/yyp/DATA/Public_Datasets/COCO/2014/train2014/'
savePath="/media/yyp/DATA/Public_Datasets/COCO/2014/JPEGImages/"
count=0
last_id = 21912
nameStr = nameStr[21911:]
for file in nameStr:
    img=cv2.imread(path+file)
    cv2.imwrite(savePath+file,img)
    count=count+1
    print('num: '+count.__str__()+'     '+file+'\n')
```
