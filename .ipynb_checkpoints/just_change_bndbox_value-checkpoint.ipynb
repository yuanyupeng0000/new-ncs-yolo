```python
import os, datetime
import xml.etree.ElementTree as ET
time_str = (datetime.datetime.now()).strftime("%Y%m%d")
SEED_XML_DIR = '/media/yyp/DATA/20190801_xmls_merged/'
GENE_XML_DIR = '/media/yyp/DATA/' + time_str + '_generated_xml/'
os.system('mkdir -p ' + GENE_XML_DIR)

seed_xml_names = os.listdir(SEED_XML_DIR)
seed_xml_names.sort(reverse=False)

for xml_name in seed_xml_names:
    print(xml_name)
    in_file = open(SEED_XML_DIR + xml_name)
    tree=ET.parse(in_file)
    root = tree.getroot()
    size = root.find('size')
    width = int(size.find('width').text)
    height = int(size.find('height').text)

    for obj in root.iter('object'):
        cls = obj.find('name').text
        if(cls == 'plate'):
            bndbox = obj.find('bndbox')
            #print(bndbox.find('xmin').text)
            bndbox.find('xmin').text = str(max(int(bndbox.find('xmin').text) - 3, 0))
            #print(bndbox.find('xmin').text)

            #print(bndbox.find('ymin').text)
            bndbox.find('ymin').text = str(max(int(bndbox.find('ymin').text) - 1, 0))
            #print(bndbox.find('ymin').text)

            #print(bndbox.find('xmax').text)
            bndbox.find('xmax').text = str(min(int(bndbox.find('xmax').text) + 3, width))
            #print(bndbox.find('xmax').text)

            #print(bndbox.find('ymax').text)
            bndbox.find('ymax').text = str(min(int(bndbox.find('ymax').text) + 5, height))
            #print(bndbox.find('ymax').text)

    tree.write(GENE_XML_DIR + xml_name)            
    print('xml {0} changed'.format(xml_name))
```
