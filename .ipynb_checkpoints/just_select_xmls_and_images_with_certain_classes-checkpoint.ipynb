```python
import os,datetime
import cv2
import xml.etree.ElementTree as ET
from PIL import Image
import numpy as np
import random
from pathlib import Path

classes = ['bus', 'car','truck', 'person', 'motorbike', 'bicycle']
time_str = (datetime.datetime.now()).strftime("%Y%m%d")
#image_dir = '/data/darknet/python/' + 'coco_image_rider/'
#xml_dir = '/data/darknet/python/' + '20190514coco_xml_rider/'
image_dir = '/media/yyp/DATA/Public_Datasets/ZENITH_DATASETS/VOCdevkit_motorbike_bicycle/VOC2007/JPEGImages/'
xml_dir = '/media/yyp/DATA/Public_Datasets/ZENITH_DATASETS/VOCdevkit_motorbike_bicycle/VOC2007/Annotations/'
#GENE_IMG_DIR = '/data/darknet/python/' + time_str + '_image_selected/'
#GENE_XML_DIR = '/data/darknet/python/' + time_str + '_xml_selected/'
GENE_IMG_DIR = '/data/darknet/python/' + 'test_image/'
GENE_XML_DIR = '/data/darknet/python/' + 'test_xml/'
SEED_XML_DIR = xml_dir
SEED_IMG_DIR = image_dir

#GENE_OBJ_DIR = '/media/yyp/DATA/Public_Datasets/COCO/2014/' + time_str + '_jenerated_obj/'
os.system('mkdir -p ' + GENE_IMG_DIR)
os.system('mkdir -p ' + GENE_XML_DIR)
'''
for i in classes:
    os.system('mkdir ' + selected_image_dir + i)
    os.system('mkdir ' + selected_xml_dir + i)
'''
    
def IOU(bbox_a, bbox_b):
    '''
    W = min(A.RT.x, B.RT.x) - max(A.LB.x, B.LB.x) 
    H = min(A.RT.y, B.RT.y) - max(A.LB.y, B.LB.y) 
    if W <= 0 or H <= 0: 
        return 0 
    SA = (A.RT.x - A.LB.x) * (A.RT.y - A.LB.y) 
    SB = (B.RT.x - B.LB.x) * (B.RT.y - B.LB.y) 
    cross = W * H return cross/(SA + SB - cross)
    '''
    W = min(bbox_a[1], bbox_b[1]) - max(bbox_a[0], bbox_b[0]) 
    H = min(bbox_a[3], bbox_b[3]) - max(bbox_a[2], bbox_b[2]) 
    if W <= 0 or H <= 0: 
        return 0
    SA = (bbox_a[1] - bbox_a[0]) * (bbox_a[3] - bbox_a[2]) 
    SB = (bbox_b[1] - bbox_b[0]) * (bbox_b[3] - bbox_b[2])  
    cross = W * H 
    return cross/(SA + SB - cross)

def static_judgement_include(bbox_bigger, bbox_smaller):
    if(bbox_bigger[0] <= bbox_smaller[0] and bbox_bigger[1] >= bbox_smaller[1] and 
       bbox_bigger[2] <= bbox_smaller[2] and bbox_bigger[3] >= bbox_smaller[3]):
        return True       
    else:
        return False
def get_all_inner_obj(objbase, objs):
    inner_objs = []
    bbox_base = get_bbox_from_xmlobj(objbase)
    for k in objs:
        bbox_k = get_bbox_from_xmlobj(k)
        result = static_judgement_include(bbox_base, bbox_k)
        if(result == True and objbase != k):
            inner_objs.append(k)
    return inner_objs
    

def get_obj_from_xml(xml):
    in_file = open(xml)   
    tree=ET.parse(in_file)
    root = tree.getroot()
    return [obj for obj in root.iter('object')]

def get_obj_from_xml_tree(xml_tree):   
    root = xml_tree.getroot()
    objs = root.findall('object')
    return objs

def get_obj_from_image_file(file, bbox):
    img = cv2.imread(file)
    img_obj = img[int(bbox[2]):int(bbox[3]), int(bbox[0]):int(bbox[1])]
    return img_obj
def get_obj_from_image_data(image_data, bbox):
    img_obj = image_data[int(bbox[2]):int(bbox[3]), int(bbox[0]):int(bbox[1])]
    return img_obj
def past_region_to_image_data(image_data, image_obj_data, bbox):
    image_data[int(bbox[2]):int(bbox[3]), int(bbox[0]):int(bbox[1])] = image_obj_data
    return image_data

def get_cls_from_xmlobj(obj_element):    
    return obj_element.find('name').text

def get_bbox_from_xmlobj(obj_element):
    xmlbox = obj_element.find('bndbox')
    return [float(xmlbox.find('xmin').text), float(xmlbox.find('xmax').text), 
            float(xmlbox.find('ymin').text), float(xmlbox.find('ymax').text)]

def get_bboxes_from_etree(etree):
    root = etree.getroot()  
    objects = root.findall('object')
    bboxes = []
    for obj in objects:
        xmlbox = obj.find('bndbox')
        b = [float(xmlbox.find('xmin').text), float(xmlbox.find('xmax').text), float(xmlbox.find('ymin').text), float(xmlbox.find('ymax').text)]
        bboxes.append(b)
    return bboxes
def get_objs_by_cls_name(objs, cls_name):
    target_objs = []
    for i in objs:
        cls = get_cls_from_xmlobj(i)
        if(cls == cls_name):
            target_objs.append(i)
    return target_objs
def remove_objs_from_etree(xmlobjs, etree):
    _root = etree.getroot()
    for obj in xmlobjs:
        _root.remove(obj)
    return etree
def remove_xmlobjs_and_blur_image_target_region(xmlobjs, etree, image_data):
    _root = etree.getroot()
    for obj in xmlobjs:
        _root.remove(obj)
        _bbox = get_bbox_from_xmlobj(obj)
        img_obj = get_obj_from_image_data(image_data, _bbox)
        
        img_obj = cv2.blur(img_obj, (100, 100))
        img_obj = cv2.blur(img_obj, (100, 100))
        image_data = past_region_to_image_data(image_data, img_obj, _bbox)
    return etree, image_data
      
seed_xml_names = os.listdir(SEED_XML_DIR)
seed_xml_names.sort()
#classes = ['bus', 'car', 'truck', 'motorbike', 'bicycle', 'person']
```

```python
##################################################################################################################
## This section for remove xmlobjs except certain class and then blur image corrosppending regions
##################################################################################################################
target_cls_name = ['rider-mo', 'rider-bi']
for xml_name in seed_xml_names:
    #print(len(xml_name))
    if(len(xml_name) != 11 and len(xml_name) != 15):
        continue
    img_name = SEED_IMG_DIR + xml_name[:-3] + 'jpg'
    xml_name = SEED_XML_DIR + xml_name
    
    print(img_name)
    print(xml_name)
    etree=ET.parse(xml_name)
    
    image_data = cv2.imread(img_name)
    img = image_data.copy()
    print(image_data.shape)
    #objs = get_obj_from_xml(xml_name)
    objs_inference = get_obj_from_xml_tree(etree)    
    bboxes = get_bboxes_from_etree(etree)
    all_found_objs = []    
    for c in target_cls_name:
        found_objs = get_objs_by_cls_name(objs_inference, c)
        all_found_objs += found_objs
    all_inner_objs = []
    for j in all_found_objs:
        inner_objs = get_all_inner_obj(j, objs_inference)
        all_inner_objs += inner_objs
    left_all_objs = []
    for k in objs_inference:
        if(k not in all_inner_objs and k not in all_found_objs):
            left_all_objs.append(k)
    if(len(left_all_objs) == 0):
        continue
    '''
    print('objs_inference:{0}'.format(objs_inference))
    print('all_found_objs:{0}'.format(all_found_objs))
    print('all_inner_objs:{0}'.format(all_inner_objs))
    print('left_all_objs:{0}'.format(left_all_objs))
    print('------------------------------------------')
    '''
    objs_motorbike = get_objs_by_cls_name(left_all_objs, 'motorbike')
    objs_person = get_objs_by_cls_name(left_all_objs, 'person')
    objs_bicycle = get_objs_by_cls_name(left_all_objs, 'bicycle')
    objs_should_keep = []
    for person in objs_person:
        bbox = get_bbox_from_xmlobj(person)
        if(((bbox[3] - bbox[2])/(bbox[1] - bbox[0]) >= 2.5) or (((bbox[3] - bbox[2])/(bbox[1] - bbox[0]) > 1.8) and (bbox[3] - bbox[2]) > 250)):
            objs_person.remove(person)
            objs_should_keep.append(person)
    #break
    #etree, image_data = remove_xmlobjs_and_blur_image_target_region(left_all_objs, tree, img)
    etree, image_data = remove_xmlobjs_and_blur_image_target_region(objs_person, etree, image_data)
    etree, image_data = remove_xmlobjs_and_blur_image_target_region(objs_motorbike, etree, image_data)
    etree, image_data = remove_xmlobjs_and_blur_image_target_region(objs_bicycle, etree, image_data)
    
    #fix the image obj caused by up blur(condation: bboxes overlap)
    objs_should_keep = objs_should_keep + all_found_objs
    for i in objs_should_keep:
        bbox = get_bbox_from_xmlobj(i)
        image_obj_data = get_obj_from_image_data(img, bbox)
        image_data = past_region_to_image_data(image_data, image_obj_data, bbox)
        
    found_objs = get_objs_by_cls_name(left_all_objs, 'delete')
    etree = remove_objs_from_etree(found_objs, etree)
    found_objs = get_objs_by_cls_name(left_all_objs, 'good')
    etree = remove_objs_from_etree(found_objs, etree)
    cv2.imwrite(GENE_IMG_DIR  + img_name.split('/')[-1], image_data)
    etree.write(GENE_XML_DIR  + xml_name.split('/')[-1])
    #break
```
